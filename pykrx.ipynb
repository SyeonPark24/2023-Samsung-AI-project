```python
# 데이터프레임으로 하기 위해서 판다스 이용
import pandas as pd
# 버전 확인
pd.__version__
```

```python
# 상장주식수가 겹치기 때문에 제거해주기
nc1.drop('상장주식수',axis = 1, inplace = True)
netm1.drop('상장주식수',axis = 1, inplace = True)
craft1.drop('상장주식수',axis = 1, inplace = True)
sselec1.drop('상장주식수',axis = 1, inplace = True)
skhy1.drop('상장주식수',axis = 1, inplace = True)
dbhy1.drop('상장주식수',axis = 1, inplace = True)
hcar1.drop('상장주식수',axis = 1, inplace = True)
hmo1.drop('상장주식수',axis = 1, inplace = True)
hlmd1.drop('상장주식수',axis = 1, inplace = True)
kia1.drop('상장주식수',axis = 1, inplace = True)
lgino1.drop('상장주식수',axis = 1, inplace = True)
kt1.drop('상장주식수',axis = 1, inplace = True)
lgen1.drop('상장주식수',axis = 1, inplace = True)
sssdi1.drop('상장주식수',axis = 1, inplace = True)
lgchem1.drop('상장주식수',axis = 1, inplace = True)
skinno1.drop('상장주식수',axis = 1, inplace = True)
```

```python
# 기업들을 하나의 행으로 합쳐주기
nc2 = pd.concat([nc, nc1], axis = 1)
netm2= pd.concat([netm, netm1], axis = 1)
craft2= pd.concat([craft, craft1], axis = 1)
sselec2= pd.concat([sselec, sselec1], axis = 1)
skhy2= pd.concat([skhy, skhy1], axis = 1)
dbhy2= pd.concat([dbhy, dbhy1], axis = 1)
hcar2= pd.concat([hcar, hcar1], axis = 1)
hmo2= pd.concat([hmo, hmo1], axis = 1)
hlmd2= pd.concat([hlmd, hlmd1], axis = 1)
kia2= pd.concat([kia, kia1], axis = 1)
lgino2= pd.concat([lgino, lgino1], axis = 1)
kt2= pd.concat([kt, kt1], axis = 1)
lgen2= pd.concat([lgen, lgen1], axis = 1)
sssdi2= pd.concat([sssdi, sssdi1], axis = 1)
lgchem2= pd.concat([lgchem, lgchem1], axis = 1)
skinno2= pd.concat([skinno, skinno1], axis = 1)
```

```python
# 모든 기업들을 하나의 데이터 프레임으로 합치기
result = pd.concat([nc2, netm2, craft2, sselec2,skhy2,dbhy2,hcar2,hmo2,hlmd2,kia2,lgino2,kt2,lgen2,sssdi2,lgchem2,skinno2], axis = 0)
```

```python
# 데이터프레임에 기업 코드 넣어주기
result['code'] = ['036570','251270','259960','005930','000660','000990','005380','012330','204320','000270','011070','030200','373220','006400','051910','096770']
```

```python
# 인덱스가 날짜이므로 제거해주기
result = result.reset_index(drop=True)
```
